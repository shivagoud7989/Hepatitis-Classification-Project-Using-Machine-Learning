PREPARING TRAINING AND TESTING DATA

```python
dummy = pd.get_dummies(data,drop_first = True)
dummy
```

```python
dummy.columns
```

```python
p_cols = [' AGE', ' SEX', ' STEROID', ' ANTIVIRALS', ' FATIGUE',
       ' MALAISE', ' ANOREXIA', ' LIVER BIG', ' LIVER FIRM',
       ' SPLEEN PALPABLE', ' SPIDERS', ' ASCITES', ' VARICES', ' BILIRUBIN',
       ' ALK PHOSPHATE', ' SGOT', ' ALBUMIN', ' PROTIME', ' HISTOLOGY']
X=dummy[p_cols]
y=dummy['Class']
```

**MODEL BUILDING**

LOGISTIC REGRESSION

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# Assume you have a dataset with features X and labels y
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=3)
X_train.shape,X_test.shape
```

```python
# Create a logistic regression model
logreg = LogisticRegression()

# Fit the model to the training data
logreg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = logreg.predict(X_test)

# Evaluate the model
print("Accuracy: ",metrics.accuracy_score(y_test, y_pred))
```

K NEAREST NEIGHBORS

```python
# Import necessary libraries
from sklearn.neighbors import KNeighborsClassifier

# Create a KNN model with k=3 (you can adjust this parameter)
knn = KNeighborsClassifier(n_neighbors=3)

# Fit the model to the training data
knn.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test)
y_pred
```

```python
# Evaluate the model
print("Accuracy: ",metrics.accuracy_score(y_test, y_pred))
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred
```

```python
# Evaluate the model
print("Accuracy: ",metrics.accuracy_score(y_test, y_pred))
```

NAIVE BAYES

```python
from sklearn.naive_bayes import GaussianNB

# Create a Gaussian Naive Bayes classifier
nb = GaussianNB()
nb.fit(X_train, y_train)

y_pred = nb.predict(X_test)
y_pred
```

```python
# Evaluate the model
print("Accuracy: ",metrics.accuracy_score(y_test, y_pred))
```

SUPPORT VECTOR MACHINE

```python
from sklearn.svm import SVC

# Create a linear SVM classifier
clf = SVC(kernel='linear')
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
y_pred
```

```python
# Evaluate the model
print("Accuracy: ",metrics.accuracy_score(y_test, y_pred))
```

NEURAL NETWORK

```python
# Import necessary libraries
import tensorflow as tf
from sklearn.metrics import accuracy_score, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Build the neural network model
nn = Sequential()
nn.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
nn.add(Dense(32, activation='relu'))
nn.add(Dense(1, activation='sigmoid'))

# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Train the model
nn.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
# Make predictions on the test set
y_pred_proba = nn.predict(X_test)
y_pred = (y_pred_proba > 0.5).astype(int)
```

```python
# Evaluate the model
print("Accuracy: ",metrics.accuracy_score(y_test, y_pred))
```

BAGGING

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

# Create a Decision Tree base classifier
base_classifier = DecisionTreeClassifier()

# Create a Bagging Classifier with the base classifier
bagging_classifier = BaggingClassifier(base_classifier, n_estimators=10, random_state=0)
bagging_classifier.fit(X_train, y_train)

y_pred = bagging_classifier.predict(X_test)
y_pred
```

```python
# Evaluate the model
print("Accuracy: ",metrics.accuracy_score(y_test, y_pred))
```

GRADIENT BOOSTING

```python
from sklearn.ensemble import GradientBoostingClassifier

# Create a Gradient Boosting Classifier with desired hyperparameters
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=0)
gbc.fit(X_train, y_train)

y_pred = gbc.predict(X_test)
y_pred
```

```python
# Evaluate the model
print("Accuracy: ",metrics.accuracy_score(y_test, y_pred))
```

MULTICOLLINEARITY

```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```
